## Model Training

### Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
```

Which features should be good in which positions were determined using the a apriori algorithm.

```python
def feature_selection(position, side, feature_threshold=10, min_support_threshold = 0.8):
    if position == "GK":
        features = Goolkeeping + Mental + Physical + Other_Feature
    else:
        features = Technical + Mental + Physical + Other_Feature
    
    while min_support_threshold >= 0:
        data = df[(df['Best_Position'] == position) & (df['Side'] == side)][features] > feature_threshold
        frequent_itemsets = apriori(data, min_support=min_support_threshold, use_colnames=True)
        frequent_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 10)]
        frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)
        frequent_itemsets = frequent_itemsets.reset_index(drop=True)
        
        if len(frequent_itemsets) > 0:
            return frequent_itemsets.itemsets[0]
        else:
            min_support_threshold -= 0.05  # Decrease threshold

    return None  # Return None if no features are found
```

```python
def features_and_players(position, side):
    important_features = feature_selection(position, side)

    selected_columns = [col for col in list(df.columns) if col not in All_features[0]] + list(important_features)
    result_df = df[(df['Best_Position'] == position) & (df['Side'] == side)][selected_columns]
    
    return important_features, result_df
```

### Content-Based Filtering
Use player attributes and characteristics to recommend similar players.

```python
def recommend_similar_players(df, player_name, top_n=5):
    
    # Ensure the player_name exists in the DataFrame
    if player_name not in df['Name'].values:
        raise ValueError("Player name not found in the DataFrame")

    # Find the index of the player
    player_index = df[df['Name'] == player_name].index[0]

    # Specify the important features
    features = list(important_features) + ["Age", "ca"]
    
    # Select only the relevant features
    features_df = df[features]

    # Compute cosine similarity between players
    similarity_matrix = cosine_similarity(features_df)

    # Convert to DataFrame for easier handling
    similarity_df = pd.DataFrame(similarity_matrix, index=df.index, columns=df.index)

    # Get the top N similar players for the specified player_index
    similar_players_indices = similarity_df[player_index].sort_values(ascending=False)[1:top_n+1].index

    # Create a DataFrame with the recommended players and their similarity scores
    recommendations = df.loc[similar_players_indices].copy()
    recommendations['Cosine Similarity'] = similarity_df.loc[similar_players_indices, player_index].values

    return recommendations
```

```python
similar_players = recommend_similar_players(df_position, "Martin √òdegaard")
display(similar_players)
```

### Diversity and Novelty

```python
def evaluate_diversity_and_novelty(df, recommend_function, top_n=5):
    # Counter for keeping track of how many times each player is recommended
    recommendation_counts = Counter()

    # Loop over each player in the DataFrame
    for player in df['Name']:
        # Get top N recommendations for each player
        recommended_players = recommend_function(df, player, top_n)

        # Update counts
        recommendation_counts.update(recommended_players['Name'])

    # Convert the counter to a DataFrame for better visualization
    recommendation_freq_df = pd.DataFrame(recommendation_counts.items(), columns=['Player', 'Recommendation_Count'])
    recommendation_freq_df = recommendation_freq_df.sort_values(by='Recommendation_Count', ascending=False)

    return recommendation_freq_df
```

```python
diversity_novelty_df = evaluate_diversity_and_novelty(df_position, recommend_similar_players)
display(diversity_novelty_df)
```
